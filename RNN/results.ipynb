# confusion matrix

## conf matrix single det

```python
from dataset import LABELS, TensorflowDataset
import os
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

label_detection_validation = TensorflowDataset('extracted_features\\features_validation.pkl', labels=LABELS).dataset
x = pd.read_pickle('extracted_features\\features_validation.pkl')
X = [x[0] for x in x]
y = [x[1] for x in x]
LABELSdict = {}
for i, label in enumerate(LABELS):
    LABELSdict[i] = label
LABELSdict[10] = 'unknown'
for i, value in enumerate(y):
    if value in LABELSdict.values():
        y[i] = value
    else:
        y[i] = 'unknown'
import tensorflow as tf

y_ = pd.DataFrame(y)

def conf_matrix(model_name):
    model = tf.keras.models.load_model(f'models/FINAL/TEST_{model_name}_simple_predict.h5')

    lstm_simp = np.argmax(model.predict(label_detection_validation.batch(32)), axis=1)
    lstm_simp = [LABELSdict[x] for x in lstm_simp]
    
    xaxis_cols = []
    for i, value in enumerate(y):
        if len(xaxis_cols) == 11:
            break
        if value not in xaxis_cols:
            xaxis_cols.append(value)
    
    lstm_simp_ = pd.DataFrame(lstm_simp)
    lstm__ = lstm_simp_[y_[0] != 'unknown']
    lstm_ = lstm_simp_[y_[0] == 'unknown']

    # First confusion matrix
    cm1 = confusion_matrix(y_[y_[0] != 'unknown'], lstm__, labels=xaxis_cols)
    df_cm1 = pd.DataFrame(cm1, index=xaxis_cols, columns=xaxis_cols)

    # Second confusion matrix
    cm2 = confusion_matrix(y_[y_[0] == 'unknown'], lstm_, labels=xaxis_cols)
    df_cm2 = pd.DataFrame(cm2, index=xaxis_cols, columns=xaxis_cols)

    # Create figure with two subplots
    fig, axs = plt.subplots(ncols=2, figsize=(20,7))

    # Plot first heatmap
    sns.heatmap(df_cm1, annot=True, fmt='.0f', cmap='Blues', ax=axs[0])
    axs[0].set_xlabel('Predicted')
    axs[0].set_ylabel('True')
    axs[0].set_title(f'{model_name.upper()} simple - without unknown')

    # Plot second heatmap
    sns.heatmap(df_cm2, annot=True, fmt='.0f', cmap='Blues', ax=axs[1])
    axs[1].set_xlabel('Predicted')
    axs[1].set_ylabel('True')
    axs[1].set_title(f'{model_name.upper()} simple - only unknown')

    plt.show()

```

```python
conf_matrix('lstm')
```

```python
conf_matrix('gru')
```

## conf matrix unknown det

```python
from dataset import LABELS, TensorflowDataset_unknown
import os
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

label_detection_validation = TensorflowDataset_unknown('extracted_features\\features_validation.pkl', labels=LABELS).dataset
x = pd.read_pickle('extracted_features\\features_validation.pkl')
X = [x[0] for x in x]
y = [x[1] for x in x]
LABELSdict = {}
for i, label in enumerate(LABELS):
    LABELSdict[i] = label
for i, value in enumerate(y):
    if value in LABELSdict.values():
        y[i] = 'known'
    else:
        y[i] = 'unknown'
import tensorflow as tf

def conf_matrix(model_name = 'lstm'):
    model = tf.keras.models.load_model(f'models/FINAL/TEST_{model_name}_known.h5')

    lstm_simp = np.argmax(model.predict(label_detection_validation.batch(32)), axis=1)

    lstm_simp = ['unknown' if x == 1 else 'known' for x in lstm_simp]
    xaxis_cols = []
    for i, value in enumerate(y):
        if len(xaxis_cols) == 2:
            break
        if value not in xaxis_cols:
            xaxis_cols.append(value)



    # First confusion matrix
    cm1 = confusion_matrix(y, lstm_simp, labels=xaxis_cols)
    df_cm1 = pd.DataFrame(cm1, index=xaxis_cols, columns=xaxis_cols)

    # Create figure with two subplots
    fig, axs = plt.subplots(ncols=1, figsize=(10,7))

    # Plot first heatmap
    sns.heatmap(df_cm1, annot=True, fmt='.0f', cmap='Blues', ax=axs)
    axs.set_xlabel('Predicted')
    axs.set_ylabel('True')
    axs.set_title(f'{model_name.upper()} predict - known/unknown')

```

## conf matrix only labels

```python
from dataset import LABELS, TensorflowDataset_labels
import os
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

label_detection_validation = TensorflowDataset_labels('extracted_features\\features_validation.pkl', labels=LABELS).dataset
x = pd.read_pickle('extracted_features\\features_validation.pkl')
X = [x[0] for x in x]
y = [x[1] for x in x]
LABELSdict = {}
for i, label in enumerate(LABELS):
    LABELSdict[i] = label
new = []
for i, value in enumerate(y):
    if value in LABELSdict.values():
        new.append(value)
y = new

import tensorflow as tf

y_ = pd.DataFrame(y)

def conf_matrix(model_name):
    model = tf.keras.models.load_model(f'models/FINAL/TEST_{model_name}_labels_only.h5')

    lstm_simp = np.argmax(model.predict(label_detection_validation.batch(32)), axis=1)

    lstm_simp = [LABELSdict[x] for x in lstm_simp]

    xaxis_cols = []
    for i, value in enumerate(y):
        if len(xaxis_cols) == 10:
            break
        if value not in xaxis_cols:
            xaxis_cols.append(value)



    # First confusion matrix
    cm1 = confusion_matrix(y, lstm_simp, labels=xaxis_cols)
    df_cm1 = pd.DataFrame(cm1, index=xaxis_cols, columns=xaxis_cols)

    # Create figure with two subplots
    fig, axs = plt.subplots(ncols=1, figsize=(10,7))

    # Plot first heatmap
    sns.heatmap(df_cm1, annot=True, fmt='.0f', cmap='Blues', ax=axs)
    axs.set_xlabel('Predicted')
    axs.set_ylabel('True')
    axs.set_title(f'{model_name.upper()} predict labels')

```
